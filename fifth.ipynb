```python
import numpy as np

# Original array
arr = np.array([10, 20, 30, 40, 50])

# Create a view
b = arr.view()

print("Original array:", arr)
print("View array:", b)

# Modify the view
b[2] = 99

print("\nAfter modifying the view:")
print("Original array:", arr)
print("View array:", b)
```

```python
import numpy as np

# Original array
arr = np.array([10, 20, 30, 40, 50])

# Create a copy
c = arr.copy()

print("Original array:", arr)
print("Copied array:", c)

# Modify the copy
c[2] = 99

print("\nAfter modifying the copy:")
print("Original array:", arr)
print("Copied array:", c)
```

Slicing

Slicing With List

```python
a = [1,2,3,4,[5,6,7],8,9]
print(a[0:4])

print(a[4][1])
```

Slicing With Array

```python
import numpy as np
```

```python
arr = np.array([1,2,4,5,6,7,7,8,8])
arr[2]
```

```python
arr = np.array([[1,2,4,5],[6,7,7,8]])
print(arr[0,0])
```

```python
print(arr[0:2,0])
```

```python
print(arr[0:2,0:2])
```

Fancy Indexing

```python
a = np.array([1,2,3,4,6,7,8])
b = [1,3,4]
a[b]
```

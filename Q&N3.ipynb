# Create an empty array of size 5.


```python
import numpy as np

arr = np.empty(5)
print(arr)
```

# Create an array filled with 7.

```python
arr = np.full((3,3), 7)
print(arr)
```

# Create an array of even numbers from 2 to 20.

```python
arr = np.arange(1,20,2)
print(arr)
```

# Create a 4×4 matrix of zeros.

```python
arr = np.zeros((4,4))
print(arr)
```

```python
import time 
n= 10

#lists
list1 = list(range(n))
list2 = list(range(n))

start = time.time()
result_list = [list1[i]+list2[i] for i in range(n)]
list_time = time.time() - start


# numpy arrays
arr1 = np.arange(n)
arr2 = np.arange(n)

start = time.time()
result_arr = arr1 + arr2
numpy_time = time.time() - start

print(f"list time : {list_time:.6f} sec")
print(f"numpy time: {numpy_time:.6f} sec")
print(f"Numpy is apporaximately {list_time/numpy_time:.2f}x faster")

```

Create a NumPy array containing 10, 20, 30, 40, 50.

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
print(arr)
```

Create an array of zeros.

```python
arr = np.zeros(5)
print(arr)
```

Create an array of ones.

```python
arr = np.ones((3, 4))
print(arr)
```

Create an array using arange().

```python
arr = np.arange(5, 30, 5)
print(arr)
```

Create evenly spaced numbers.

```python
arr = np.linspace(0, 20, 5)
print(arr)
```

Find the shape of an array.

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr.shape)
```

Find dimensions

```python

print(arr.ndim)
```

Randomly choose an element.

```python
arr = np.array([10,20,30,40,50])
print(np.random.choice(arr))
```
